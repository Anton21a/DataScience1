```python
import pandas as pd
df = pd.read_csv("clean_data.csv")
df
```

```python
import statsmodels.formula.api as smf

df["happiness"] = df["happiness"].astype("int")
df["financial_satisfaction"] = df["financial_satisfaction"].astype("category")
df["religion"] = df["religion"].astype("category")
df["gender"] = df["gender"].astype("category")

model = smf.ols("happiness ~ financial_satisfaction + age + gender + religion", data=df).fit()

model.summary()
```
